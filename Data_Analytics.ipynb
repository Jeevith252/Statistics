### **REMOVAL OF OUTLIERS AND CREATING BOXPLOT FOR THE DATA**

```python
import numpy as np
arr=np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
```

```python
#Measuring Q1,Q3 and IQR

Q1=np.percentile(arr,25)
Q3=np.percentile(arr,75)
IQR =Q3 - Q1
print("Q1 is:",Q1,end="\n")
print("Q3 is:",Q3)
print("IQR is:",IQR)
```

```python
#Calculating Upper Fence and Lower Fence

UF=Q3 + (1.5*IQR)
LF=Q1 - (1.5*IQR)
print("Upper Fence is:",UF)
print("Lower Fence is:",LF)
```

```python
#Removing Outliers

l=[]
for i in arr:
  if i <=UF and i >= LF:
    l.append(i)
arr2=np.array(l)
```

```python
#BOXPLOT

import seaborn as sns

sns.boxplot(x=arr2)
```

```python
sns.boxplot(x=arr)
```

#**Z- Test**

```python
import numpy as np
from scipy.stats import norm
```

```python
sample=[172,174,168,169,171,173,175,170,169,172]
population_mean=170
population_std=3
sample_mean=np.mean(sample)
n=len(sample)
```

```python
z_score=(sample_mean-population_mean)/(population_std/np.sqrt(n))
```

```python
p_value=2*(1-norm.cdf(p_value))
alpha=0.05
```

```python
if p_value < alpha:
  print("Rejected Null Hypothesis")
else:
  print("Accepted Null Hypothesis")
```

# **T-Test**

```python
import numpy as np
from scipy import stats
```

```python
sample=[172,174,168,169,171,173,175,170,169,172]
population_mean=170 #given
```

```python
sample_mean=np.mean(sample)
sample_std=np.std(sample,ddof=1)
n=len(sample)
DF=n-1
```

```python
t_stats=(sample_mean - population_mean) /(sample_std/np.sqrt(n))
```

```python
p=2*(1 - stats.t.cdf(abs(t_stats),df=n-1))
alpha=0.05
```

```python
if p<alpha:
  print("Reject Null hypothesis")
else:
  print("Accept Null hypothesis")
```

# **Two sample T-Test**

```python
import numpy as np
from scipy import stats
```

```python
g1=[85,88,90,92,87,85,89,91,86,88]
g2=[82,84,80,83,81,79,78,85,84,83]
```

```python
t_stats,p_value=stats.ttest_ind(g1,g2,equal_var=False)
alpha=0.05
```

```python
if p_value<alpha:
  print("Reject Null hypothesis")
else:
  print("Accept Null hypothesis")
```

# **CHI_SQUARE TEST**

```python
import pandas as pd
import seaborn as sns
from scipy.stats import chi2_contingency
```

```python
df=sns.load_dataset('titanic')
df
```

```python
contingency_table = pd.crosstab(df['sex'] , df['survived'])
contingency_table
```

```python
chi2, p, dof, expected = chi2_contingency(contingency_table)
alpha=0.05
```

```python
if p < alpha:
  print("Reject")
else:
  print("Accept")
```
